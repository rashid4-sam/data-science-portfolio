```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r'C:\Users\admin\Documents\All Codes\Regression\boston.csv')
```

```python
df.head()
```

```python
df.drop(columns=['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#Splitting independent and target variable
X = df.drop(columns = ['Price'],axis=1)
y = df['Price']
```

```python
#Splitting into Train and test

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
len(df)
```

```python
#Apply Linear Regression 
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred_lr = model.predict(X_test)
r2_score_lr = r2_score(y_test,y_pred_lr)
print("R2 Score of Linear Regression : ",r2_score_lr)
```

Ridge && Lasso

```python
#Applying  Ridge regression
from sklearn.linear_model import Ridge,Lasso
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train,y_train)
y_pred_rdg = ridge_model.predict(X_test)
r2_score_rdg = r2_score(y_test,y_pred_rdg)
print("R2 Score of Ridge Regression of alpha 1.0 : ",r2_score_rdg)
```

```python
#Applying  Ridge regression
from sklearn.linear_model import Ridge,Lasso
ridge_model = Ridge(alpha=0.5)
ridge_model.fit(X_train,y_train)
y_pred_rdg = ridge_model.predict(X_test)
r2_score_rdg = r2_score(y_test,y_pred_rdg)
print("R2 Score of Ridge Regression of alpha 0.5 : ",r2_score_rdg)
```

```python
lasso_reg = Lasso(alpha=0.1)
lasso_reg.fit(X_train,y_train)
y_pred_lso = lasso_reg.predict(X_test)
r2_score_lso = r2_score(y_test,y_pred_lso)
print("R2 Score of Lasso Regression of alpha 0.5 : ",r2_score_lso)
```

```python
lasso_reg = Lasso(alpha=1.0)
lasso_reg.fit(X_train,y_train)
y_pred_lso = lasso_reg.predict(X_test)
r2_score_lso = r2_score(y_test,y_pred_lso)
print("R2 Score of Lasso Regression of alpha 1.0 : ",r2_score_lso)
```

```python
# Find the coefficient with bad slope results

bad_feautures = np.where(lasso_reg.coef_ == 0)[0]
print("feautures  with bad slope result",list(X.columns[bad_feautures]))
```

```python
X_train_filtered = X_train.drop(X_train.columns[bad_feautures],axis=1)
X_test_filtered = X_test.drop(X_test.columns[bad_feautures],axis=1)

#implementing Linear reg,Lasso with filtered values
#Apply Linear Regression 
model.fit(X_train_filtered,y_train)
y_pred_lr_filtered = model.predict(X_test_filtered)
r2_score_lr_filtered = r2_score(y_test,y_pred_lr_filtered)
print("R2 Score of Linear Regression with filtered value : ",r2_score_lr_filtered)

lasso_reg_filtered = Lasso(alpha=0.1)
lasso_reg_filtered.fit(X_train_filtered,y_train)
y_pred_lso_filtered = lasso_reg_filtered.predict(X_test_filtered)
r2_score_lso_filtered = r2_score(y_test,y_pred_lso_filtered)
print("R2 Score of Lasso Regression  with filtered values: ",r2_score_lso_filtered)
```

```python
small_coefficients = np.where((lasso_reg.coef_ < 0.05) & (lasso_reg.coef_ > -0.05))[0]
print("small_coefficients  with bad slope result",list(X.columns[small_coefficients]))
```

```python
X_train_small = X_train.drop(X_train.columns[small_coefficients],axis=1)
X_test_small = X_test.drop(X_test.columns[small_coefficients],axis=1)

lasso_reg_filtered_small = Lasso(alpha=0.1)
lasso_reg_filtered_small.fit(X_train_small,y_train)
y_pred_lso_filtered_small = lasso_reg_filtered_small.predict(X_test_small)
r2_score_lso_filtered_small = r2_score(y_test,y_pred_lso_filtered_small)
print("R2 Score of Lasso Regression  with filtered small values: ",r2_score_lso_filtered_small)
```

```python
from itertools import combinations

best_r2_score = 0
best_column_combination = None

# Iterate through all possible combinations of columns to drop
for r in range(1, len(small_coefficients) + 1):
    for subset in combinations(small_coefficients, r):
        # Create a new X_train with the current subset of columns dropped
        X_train_subset = X_train.drop(X_train.columns[list(subset)], axis=1)
        X_test_subset = X_test.drop(X_test.columns[list(subset)], axis=1)

        # Apply linear regression on the subset of columns
        model_subset = LinearRegression()
        model_subset.fit(X_train_subset, y_train)

        # Calculate R2 score for the subset of columns
        r2_subset = model_subset.score(X_test_subset, y_test)

        # Update the best R2 score and column combination if a higher score is achieved
        if r2_subset > best_r2_score:
            best_r2_score = r2_subset
            best_column_combination = subset

print("Best R2 score:", best_r2_score)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('CustomerChurn.csv')
```

```python
df.head(1)
```

```python
df.info()
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors='coerce')
```

```python
df.dropna(how='any',inplace=True)
```

```python
df.Churn.value_counts()/len(df)*100
```

```python
X = df.drop(['customerID','Churn'],axis=1)
y = df.Churn.values
```

```python
X.head(1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
X.columns
```

```python
#feature encode
X = pd.get_dummies(X,columns=['gender',  'Partner', 'Dependents',
       'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod'],drop_first=True).astype(int)
```

```python
X.head(1)
```

```python
#Split the data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
# We have to Feauture scale the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
X_train_sc
```

Decision Tree


```python
#Implement Dt Model
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train_sc,y_train)
```

```python
# Making the prediction using X_test values
y_pred_dt = model.predict(X_test_sc)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_dt)
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score,classification_report,roc_auc_score
precision_dt = precision_score(y_test,y_pred_dt)
recall_dt = recall_score(y_test,y_pred_dt)
f1_dt = f1_score(y_test,y_pred_dt)
cls_dt = classification_report(y_test,y_pred_dt)
ROC_AUC = roc_auc_score(y_test,y_pred_dt)
print("pre: ",precision_dt)
print("recall: ",recall_dt)
print("f1_score: ",f1_dt)
print("class_report: ",cls_dt)
print("AUC ROC: ",ROC_AUC)
```
